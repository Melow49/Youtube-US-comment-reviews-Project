```python
import pandas as pd
from sqlalchemy import create_engine


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
#open videos csv file

videos = pd.read_csv('./Resources/USvideos.csv',error_bad_lines=False)

videos.head()
```

```python
#open comments file 
comments = pd.read_csv('./Resources/UScomments.csv', error_bad_lines=False)
comments.head()
```

```python
videos['category_id'] = videos['category_id'].astype(int)
videos.head()
```

```python
#total comments by us customers
total_comment= videos["comment_total"].max()
total_comment
```

```python
#merging both file into one 
youtube_df = pd.merge(videos, comments, on="video_id")
youtube_df.head()
```

```python
#convert the date in to regular format

youtube_df['date'] =  pd.to_datetime(youtube_df.date, format="%d.%m")
youtube_df.head()
```

```python
#rename all the column names

youtube_df = youtube_df.rename(columns={"video_id":"video_id",
                                        "title": "title",
                                      "channel_title": "channel",
                                      "category_id": "category",
                                      "tags": "tags",
                                      "views": "views",
                                      "likes_x":"likes",
                                      "dislikes": "dislike",
                                      "comment_total": "total-comment",
                                      "thumbnail_link": "thumbnail-links",
                                      "date": "date",
                                      "comment_text": "text-comments",
                                      "likes_y": "likes-y",
                                      "replies": "replies"
                                       })
youtube_df.head()
```

```python
USVideo_Likes = youtube_df.sort_values('likes', ascending=False).drop_duplicates('video_id').sort_index()
USVideo_Likes.groupby('category').likes.mean().plot(kind="bar")
print(USVideo_Likes.groupby('category').likes.mean())
plt.title("Average Likes per Category")
plt.show()
```

```python
#save it to output folder
youtube_df.to_csv("./Output/youtube.csv")
```

```python
#creating csv file 1

table_1df= youtube_df[["video_id","title","channel","likes","dislike","category"]].copy()
table_1df.head()
```

```python
#save it to output folder

table_1df.to_csv("./Output/table_1df.csv")
```

```python
#creating csv file 2 for comments, tags and number of views
table_2df = youtube_df[["category", "video_id","total-comment","thumbnail-links", "views", "tags", "text-comments"]].copy()
table_2df.head()
```

```python

table_2df.to_csv("./Output/table_2df.csv")
```

```python
#separate data set for categories and number of likes

table_3df = youtube_df[["category", "text-comments","likes-y", "replies","date", "likes"]].copy()
table_3df.head()
```

```python
#save it to output folder

table_3df.to_csv("./Output/table_3df.csv")
```

```python
#connect to pdadmin 

rds_connection_string = "postgres:password@127.0.0.1/etl-db"
engine = create_engine(f'postgres://{rds_connection_string}')
```

```python
#added the table to pgadmin
table_1df.to_sql(name='youtube', con=engine, if_exists='append', index=False)
```

```python
#added 2nd table to pgadmin
table_2df.to_sql(name='comments', con=engine, if_exists='append', index=False)
```

```python
#added 3rd table to pdadmin
table_3df.to_sql(name='category', con=engine, if_exists='append', index=False)
```
